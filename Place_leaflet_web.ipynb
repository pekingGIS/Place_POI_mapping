## place areal layer

```python
# place Polygon Layer
def placeLayer_generator(place_shape,_name):
    feature_group = fol.map.FeatureGroup(name=_name)
    for i in range(0,place_shape.shape[0]):
        geometry_sample=place_shape.loc[i,'geometry']
        array=[]
        if type(geometry_sample)==multipolygon.MultiPolygon:
            for j in range(0,len(geometry_sample)):
                array.append([])
                for p in geometry_sample[j].exterior.coords:#多个环
                    array[j].append([p[1],p[0]])
        elif type(geometry_sample)==Polygon:
            for p in geometry_sample.exterior.coords:#1个环
                    array.append([p[1],p[0]])
                    
        popup_info = fol.Popup('<div style="font-size: 18pt">'+place_shape.loc[i,'region'], max_width=600)
        poly = fol.Polygon(array,fill_color='#999999',fill_opacity=0.1,color='#999999',opacity=0.2,weight=1)
        poly.add_child(popup_info)
        feature_group.add_child(poly)
    return feature_group
```

### Point Layers

```python
# pois Point Layer
def pointLayer_generator(point_shape,_name,_color):
    feature_group = fol.map.FeatureGroup(name=_name,show=False)
    for i in range(0,point_shape.shape[0]):
#         print ("point"+str(i))
        point_marker=fol.CircleMarker([point_shape.loc[i,'geometry'].y,\
                             point_shape.loc[i,'geometry'].x],\
                             radius=4,fill_color=_color,fill_opacity=0.25,weight=0)
        feature_group.add_child(point_marker)
    return feature_group
```

```python
mapa = fol.Map(location=[39.9,116.4],tiles='cartodbpositron',zoom_start=12,control_scale=True,prefer_canvas=True)
fol.TileLayer('stamenToner',show=False).add_to(mapa)
fol.TileLayer('cartodbdark_matter',show=False).add_to(mapa)

colormaps=['blue','orange','green','lightblue','purple','red','grey']
mytypelist=['Eating',
            'Housing',
            'Traffic',
            'Work',
            'Recreation',
            'Medical',
            'Outdoor'
           ]

mapa.add_child(placeLayer_generator(place_shape,'Place'))
mapa.choropleth(geo_data=place_jsontxt,
             fill_color='#555555',
             fill_opacity=0,
             line_color='#555555',
             line_opacity=0.2,
             highlight=True,
             name='Highlight hiddenlayer',
            )

for i in range(0,len(mytypelist)):
    pois_draw=pois_shape.loc[pois_shape.loc[:,'mytype']==mytypelist[i]]
    pois_draw.index=range(len(pois_draw))
#     print (i)
    mapa.add_child(pointLayer_generator(pois_draw,mytypelist[i],colormaps[i]))

mapa.add_child(fol.map.LayerControl())
mapa.save('./interactive_map_new.html')
```
